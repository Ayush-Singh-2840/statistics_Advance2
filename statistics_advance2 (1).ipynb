```python
#Question.10 Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.Generate 1,000 samples and plot the distribution. What is the expected mean and variance?
#Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.
#Objective: Understand the properties of a binomial distribution and verify them through simulation.

#Answer
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n_trials = 10  # Number of trials
p_success = 0.6  # Probability of success
n_samples = 1000  # Number of samples

# Simulate the binomial distribution
samples = np.random.binomial(n_trials, p_success, n_samples)

# Calculate expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Calculate sample mean and variance
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

# Plot the distribution
plt.hist(samples, bins=np.arange(-0.5, n_trials + 1.5, 1), density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title("Simulated Binomial Distribution (n=10, p=0.6)")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Sample Mean = {sample_mean:.2f}")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(n_trials + 1))

# Display the histogram
plt.show()

# Print the results
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")


```

```python
#Question.9 : Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail toreject the null hypothesis?
#Task: Write Python code to calculate the z-score and p-value for the given data.
#Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.


from scipy.stats import norm

# Given data
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25
significance_level = 0.05

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_dev / (sample_size**0.5))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Decision rule
reject_null = p_value < significance_level

# Display results
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
if reject_null:
    print("Decision: Reject the null hypothesis.")
else:
    print("Decision: Fail to reject the null hypothesis.")
```

# Probability Distribution, PMF, and PDF

This notebook explains the concepts of **Probability Distribution**, **Probability Mass Function (PMF)**, and **Probability Density Function (PDF)**, along with their differences.

### Probability Distribution

A **Probability Distribution** describes how the probabilities are distributed over the values of a random variable. It provides a complete description of the possible outcomes and their associated probabilities for a given random experiment.
There are two main types of probability distributions:
- **Discrete Probability Distribution**: For discrete random variables.
- **Continuous Probability Distribution**: For continuous random variables.


### Probability Mass Function (PMF)

A **Probability Mass Function (PMF)** is used for **discrete random variables**. It gives the probability that a discrete random variable takes an exact value.

Example: If you roll a fair die, the PMF would give the probability of each face appearing, like:
- \(P(X = 1) = \frac{1}{6}\), \(P(X = 2) = \frac{1}{6}\), and so on.

The sum of all probabilities in the PMF must be 1:
\[
\sum_{x} P(X = x) = 1
\]


### Probability Density Function (PDF)

A **Probability Density Function (PDF)** is used for **continuous random variables**. It gives the density of the probability that a random variable takes a particular value. Unlike PMF, the PDF does not provide the exact probability of a value (because it's continuous), but it provides the probability density.

The total area under the curve of a PDF equals 1, ensuring that the total probability is 1:
\[
\int_{-\infty}^{\infty} f(x) dx = 1
\]


### Differences Between PMF and PDF

| **Feature**               | **Probability Mass Function (PMF)**                             | **Probability Density Function (PDF)**                            |
|---------------------------|------------------------------------------------------------------|--------------------------------------------------------------------|
| **Type of Variables**      | Discrete Random Variables                                       | Continuous Random Variables                                        |
| **Output**                 | Gives the probability of exact outcomes.                        | Gives the density of the probability for a given value.            |
| **Sum vs. Integral**       | Sum of probabilities for all outcomes equals 1.                 | Integral of the PDF over the entire range equals 1.                |
| **Value Interpretation**   | The PMF gives the exact probability of a specific value.        | The PDF gives the likelihood of a value within a range, not for exact values. |
| **Example**                | Tossing a fair coin: \(P(\text{Heads}) = 0.5\)                   | The height of individuals, where the probability of getting a specific height is 0. |

#Answer

### Key Properties of the Poisson Distribution

1. **Parameter \( \lambda \)**:
   - The Poisson distribution is characterized by a single parameter \( \lambda \), which represents the **mean** and **variance** of the distribution.
   - \( \lambda \) is the average number of events occurring within a given interval.

   **Mean and Variance**:
   - Mean: \( \mu = \lambda \)
   - Variance: \( \sigma^2 = \lambda \)

2. **Probability Mass Function (PMF)**:
   The probability of observing exactly \( k \) events occurring in a fixed interval is given by the Poisson PMF:
   \[
   P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \dots
   \]
   where:
   - \( X \) is the random variable representing the number of events.
   - \( \lambda \) is the average number of events in the interval.
   - \( k! \) is the factorial of \( k \).
   - \( e \) is Euler's number (approximately 2.71828).

3. **Shape of the Distribution**:
   - For small values of \( \lambda \), the Poisson distribution is skewed.
   - As \( \lambda \) increases, the distribution becomes more symmetric and resembles a normal distribution.

4. **Memorylessness**:
   - The Poisson distribution is **memoryless**, meaning that the probability of an event occurring in the future is not affected by past events.


### When is it Appropriate to Use the Poisson Distribution?

The Poisson distribution is appropriate when:
1. **Event Occurrence is Independent**: The events occur independently of each other.
2. **Rate is Constant**: The average rate of occurrence \( \lambda \) remains constant over time or space.
3. **Discrete Events**: The events are discrete and happen one at a time.
4. **Events in a Fixed Interval**: The events happen in a fixed interval of time, space, etc.


### Real-Life Use Cases of the Poisson Distribution

1. **Call Centers**: The number of calls received by a call center per minute.
2. **Traffic Flow**: The number of cars passing through a toll booth in an hour.
3. **Natural Events**: The number of earthquakes occurring in a region over a specific period of time.
4. **Website Visits**: The number of visitors to a website during a fixed time window.
5. **Queueing Systems**: The number of customers arriving at a service counter in a specified time period.


### When Not to Use the Poisson Distribution

The Poisson distribution is not appropriate when:
1. **Events are not independent**: If the events are dependent on each other (e.g., one event increases the likelihood of the next).
2. **Variable Rate of Occurrence**: If the rate of events \( \lambda \) is not constant over time or space.




### Conclusion

The **Poisson distribution** is useful for modeling the number of events happening in a fixed interval of time or space when:
- The events are independent.
- The rate of occurrence is constant.
- The events are discrete and happen one at a time.

It is used in various fields, such as call center analysis, traffic flow modeling, and natural event occurrences.

### Conditions for Using Binomial Distribution

The **binomial distribution** is used under the following conditions:

1. **Fixed Number of Trials**: The number of trials, denoted as \( n \), is fixed.
2. **Binary Outcomes**: Each trial has two possible outcomes: success or failure.
3. **Constant Probability of Success**: The probability of success \( p \) is the same for each trial.
4. **Independence**: The trials are independent of each other.


### Relation to Bernoulli Distribution

- The **Bernoulli distribution** is a special case of the binomial distribution where the number of trials \( n = 1 \).
- In a Bernoulli distribution, there are only two possible outcomes (success or failure) in a single trial, and the probability of success is \( p \).
  
Thus, the **binomial distribution** generalizes the **Bernoulli distribution** to multiple trials, where each trial follows the Bernoulli distribution.


### Conclusion

- The **binomial distribution** is used to model the number of successes in a fixed number of independent trials.
- The **Bernoulli distribution** is a special case of the binomial distribution where there is only one trial.
- Binomial distribution generalizes Bernoulli distribution to multiple trials, with each trial having a binary outcome.

```python
#Question.2  What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?
```

#Answer

The **p-value** is a measure used in **hypothesis testing** to determine the strength of the evidence against the null hypothesis. It represents the probability of obtaining results as extreme as the observed results, or more extreme, if the null hypothesis is true.

### Hypothesis Testing Using p-value

1. **Formulate Hypotheses**: 
   - **Null Hypothesis (H₀)**: A statement that there is no effect or no difference (e.g., the population mean is equal to a specified value).
   - **Alternative Hypothesis (H₁)**: A statement that there is an effect or a difference (e.g., the population mean is not equal to the specified value).

2. **Conduct a Test**: Calculate the test statistic and find the p-value.

3. **Decision Rule**:
   - If the **p-value is less than the significance level (α)**, we **reject the null hypothesis**.
   - If the **p-value is greater than or equal to α**, we **fail to reject the null hypothesis**.


### Interpretation of p-value

- A **small p-value** (e.g., 0.01) indicates that the observed data is very unlikely under the assumption of the null hypothesis. It suggests that the null hypothesis can be rejected.
- For instance, if the p-value is 0.01, there is only a **1% probability** of observing results as extreme as those observed if the null hypothesis were true.

#### Decision Rule:
- If **p-value < α (e.g., 0.05)**: Reject the null hypothesis.
- If **p-value ≥ α (e.g., 0.05)**: Fail to reject the null hypothesis.



```python
#Question.1  Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing.
```

#Answer

### Z-Statistic

The **z-statistic** is a measure of how many standard deviations an element is from the mean. It is calculated using the formula:

\[
z = \frac{X - \mu}{\sigma}
\]

Where:
- \(X\) = Observed value or sample mean
- \(\mu\) = Population mean
- \(\sigma\) = Population standard deviation


### Relationship with Standard Normal Distribution

The **z-statistic** follows the **standard normal distribution** which has:
- Mean = 0
- Standard deviation = 1

This allows us to compare the z-statistic to the standard normal distribution to determine the probability of obtaining results as extreme as the observed ones under the null hypothesis.


### Using Z-Statistic in Hypothesis Testing

1. **Formulate Hypotheses**:
   - **Null Hypothesis (H₀)**: There is no effect or difference (e.g., the population mean is equal to 100).
   - **Alternative Hypothesis (H₁)**: There is an effect or difference (e.g., the population mean is not equal to 100).

2. **Calculate the Z-Statistic** using the formula.

3. **Compare the Z-Statistic** with the critical value from the standard normal distribution:
   - If the **absolute value of the z-statistic** is greater than the critical value, reject the null hypothesis.
   - Otherwise, fail to reject the null hypothesis.


### Conclusion of Hypothesis Testing

- **Z-Statistic** helps determine how far the sample mean is from the population mean, in terms of standard deviations.
- If the **absolute value of the z-statistic** exceeds the critical value, we **reject** the null hypothesis.
- If the **absolute value of the z-statistic** is less than the critical value, we **fail to reject** the null hypothesis.
